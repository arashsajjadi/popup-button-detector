````python
import cv2
import numpy as np
import easyocr
from ultralytics import YOLO
import subprocess
import re
import os
import matplotlib.pyplot as plt
from difflib import get_close_matches

# ---------------------------
# Global Model Initialization
# ---------------------------
yolo_model = YOLO('yolo11x.pt')  # Ensure this model is trained on popup UI elements
ocr_reader = easyocr.Reader(['en'])

# ---------------------------
# Utility Functions
# ---------------------------
def preprocess_image(image_path):
    """Read the image and convert it to RGB."""
    image = cv2.imread(image_path)
    if image is None:
        print("Error: Unable to read image.")
        return None
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image, rgb_image

def clean_text(t):
    """
    Normalize OCR output: 
    - Fix common errors (like 'not nowv' → 'not now').
    - Remove trailing artifacts.
    - Convert to lowercase.
    """
    t = t.lower().strip()
    if t.endswith('v') and len(t) > 3:
        t = t[:-1]
    if t.endswith('n') and len(t) > 3:
        t = t[:-1]
    return t

def extract_text_easyocr(image):
    """
    Run EasyOCR on the image.
    Returns a list of dictionaries with 'text' and its bounding box.
    """
    results = ocr_reader.readtext(image)
    ocr_results = []
    for bbox, text, conf in results:
        text = clean_text(text)
        if not text:
            continue
        x_coords = [pt[0] for pt in bbox]
        y_coords = [pt[1] for pt in bbox]
        left, top = int(min(x_coords)), int(min(y_coords))
        right, bottom = int(max(x_coords)), int(max(y_coords))
        ocr_results.append({
            'text': text,
            'bbox': (left, top, right, bottom),
            'conf': conf
        })
    return ocr_results

def detect_buttons_yolo(image):
    """
    Run YOLOv11x on the image and debug why it's not detecting anything.
    """
    yolo_model = YOLO("yolo11x.pt")
    results = yolo_model(image)
    img_h, img_w, _ = image.shape

    detections = []
    print(f"⚠ YOLO Raw Detections: {len(results[0].boxes)}")
    
    for result in results:
        for box in result.boxes.data.tolist():
            x1, y1, x2, y2, conf, cls = box
            
            # Print all detected bounding boxes before filtering
            print(f"📦 YOLO Detected Box: {x1:.0f}, {y1:.0f}, {x2:.0f}, {y2:.0f} | Confidence: {conf:.2f}")
            
            # Reduce filtering threshold to see if more detections show up
            if conf > 0.2:
                detections.append({
                    'bbox': (int(x1), int(y1), int(x2), int(y2)),
                    'conf': conf,
                    'class': int(cls)
                })

    print(f"✅ YOLO Filtered Detections: {len(detections)}")
    return detections


def filter_boxes_by_size(detections, img_w, img_h, min_area_ratio=0.0005, max_area_ratio=0.5):
    """
    Reject boxes that are too small or too big relative to the image size.
    """
    filtered = []
    for d in detections:
        (x1, y1, x2, y2) = d['bbox']
        w = x2 - x1
        h = y2 - y1
        area = w * h
        ratio = area / float(img_w * img_h)
        if min_area_ratio < ratio < max_area_ratio:
            filtered.append(d)
    return filtered

def draw_all_detections(orig_image, ocr_results, yolo_detections, selected_candidate=None):
    """
    Draw bounding boxes for:
    - OCR results (Blue)
    - YOLO detections (Red)
    - Final Selected Button (Yellow)
    """
    debug_image = orig_image.copy()
    
    # Draw OCR results in blue
    for ocr in ocr_results:
        (x1, y1, x2, y2) = ocr['bbox']
        cv2.rectangle(debug_image, (x1, y1), (x2, y2), (255, 0, 0), 2)
        cv2.putText(debug_image, ocr['text'], (x1, y1 - 5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
    
    # Draw YOLO results in red
    for det in yolo_detections:
        (x1, y1, x2, y2) = det['bbox']
        cv2.rectangle(debug_image, (x1, y1), (x2, y2), (0, 0, 255), 2)
    
    # Draw selected final button in yellow
    if selected_candidate:
        (x1, y1, x2, y2) = selected_candidate['bbox']
        cv2.rectangle(debug_image, (x1, y1), (x2, y2), (0, 255, 255), 3)
        cv2.putText(debug_image, selected_candidate['text'], (x1, y1 - 5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 255), 2)
    
    return debug_image

def query_llava_for_button(candidate_texts):
    """
    Query LLaVA 32B to determine which text is the close button.
    """
    prompt = f"""We have extracted the following button texts from a pop-up:
{candidate_texts}
Which one is the button that closes or dismisses the pop-up? 
Return only the exact text from the list above."""
    
    print("Querying LLaVA 32B with prompt:")
    print(prompt)
    try:
        result = subprocess.run(
            ["ollama", "run", "LLaVA:34b", prompt],
            capture_output=True,
            text=True
        )
        output = result.stdout.strip()
        m = re.search(r'"([^"]+)"', output)
        chosen_text = m.group(1) if m else output
        return chosen_text.strip().lower()
    except Exception as e:
        print("Error querying LLaVA 32B:", e)
        return None

def fuzzy_match_candidate(candidates, chosen_text):
    """
    Use fuzzy matching to find the best candidate.
    """
    candidate_texts = [cand['text'] for cand in candidates]
    matches = get_close_matches(chosen_text, candidate_texts, n=1, cutoff=0.7)
    if matches:
        for cand in candidates:
            if cand['text'] == matches[0]:
                return cand
    return None

# ---------------------------
# Main Processing Pipeline
# ---------------------------
def process_image(image_path):
    print(f"Processing image: {image_path}")
    
    orig_image, rgb_image = preprocess_image(image_path)
    if orig_image is None:
        return None

    ocr_results = extract_text_easyocr(rgb_image)
    yolo_detections = detect_buttons_yolo(rgb_image)
    
    img_h, img_w, _ = orig_image.shape
    yolo_detections = filter_boxes_by_size(yolo_detections, img_w, img_h)
    
    candidates = ocr_results  
    candidate_texts = [cand['text'] for cand in candidates]
    
    chosen_text = query_llava_for_button(candidate_texts)
    if not chosen_text:
        return None

    selected_candidate = fuzzy_match_candidate(candidates, chosen_text)
    if not selected_candidate:
        return None

    output_image = draw_all_detections(orig_image, ocr_results, yolo_detections, selected_candidate)
    output_path = "output_N_" + os.path.basename(image_path)
    cv2.imwrite(output_path, output_image)
    # Show the final image with all detections
    plt.figure(figsize=(8, 6))
    plt.imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.title(f"Final Selection: {selected_candidate['text']}")
    plt.show()
    print("Output image saved as:", output_path)

    # Calculate normalized click coordinates (center of bounding box)
    x1, y1, x2, y2 = selected_candidate['bbox']
    click_x = (x1 + x2) / 2.0
    click_y = (y1 + y2) / 2.0
    norm_x = click_x / orig_image.shape[1]
    norm_y = click_y / orig_image.shape[0]
    
    print(f"🖱 Click Coordinates: ({click_x:.2f}, {click_y:.2f}) | Normalized: ({norm_x:.2f}, {norm_y:.2f})")


    return output_path

# ---------------------------
# Main Execution
# ---------------------------
if __name__ == "__main__":
    image_paths = ["test.jpg", "test_1.png", "test_2.png", "test_3.png"]
    for img_path in image_paths:
        output_path = process_image(img_path)
        if output_path:
            print(f"Processed: {output_path}")
````

Make sure your dataset is arranged in a structure that Ultralytics expects (images and corresponding label files).

### **Step 2: Fine-Tuning with Ultralytics YOLO**

Use the following Python code to fine-tune your model:


````python
from ultralytics import YOLO

# Load a pre-trained YOLO model (e.g., 'yolo11x.pt')
model = YOLO('yolo11x.pt')

# Train the model using your custom dataset configuration.
# Adjust parameters like epochs, batch size, and image size as needed.
results = model.train(data='popups.yaml', 
                      epochs=50,         # Number of epochs (adjust as needed)
                      batch=16,          # Batch size (adjust according to your GPU memory)
                      imgsz=640,         # Image size
                      project='runs/train', 
                      name='yolo_finetuned_popups')

print("Training complete!")
````

### **Step 3: Evaluate the Fine-Tuned Model**

After training, you can evaluate the model on your validation set:

````python
# Evaluate the fine-tuned model
metrics = model.val(data='popups.yaml')
print(metrics)
````
